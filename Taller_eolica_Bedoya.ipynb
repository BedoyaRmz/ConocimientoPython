```python
##importamos time y datetime para separar las partes de la fecha
import time, datetime

def YEAR(s):
    return datetime.datetime.strptime(s, '%d/%m/%Y').strftime('%Y')

def MONTH(s):
    ## Numero del mes
    return datetime.datetime.strptime(s, '%d/%m/%Y').month

def DAYOFMONTH(s):
    return datetime.datetime.strptime(s, '%d/%m/%Y').strftime('%d')
```

```python
datetime_val = open('files/06/estacion1.csv', 'r').readlines()
datetime_val = datetime_val[1:]
datetime_val = [row[0:-1] if row[-1] == '\n' else row for row in datetime_val]
datetime_val = [row.split(';') for row in datetime_val]
datetime_Val = [[row[0], YEAR(row[0]), MONTH(row[0]),DAYOFMONTH(row[0])] for row in datetime_val]
datetime_val
```
